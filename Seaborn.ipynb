```python
import seaborn as sns
```

Seaborn using for the Data Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme()
```

```python
sns.relplot(data =tips,x = 'total_bill',y = 'tip',col = 'time',hue = 'smoker',style = 'smoker',size = 'size')
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.scatterplot(x = 'sepal_length',y = 'sepal_width',hue = 'species',data = iris)
```

```python
#Loading the Titanic Dataset ........

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(x = 'class',data = titanic)
```

```python
sns.countplot(x = 'sex',data = titanic)
```

```python
sns.barplot(x = 'sex',y = 'age',hue = 'class',data = titanic)
```

```python
sns.displot(titanic['age'])
```

```python
correlation = titanic.corr()
correlation
```

```python
#heatmap use for correlation matrix
plt.figure(figsize = (10,10))
sns.heatmap(correlation,cbar = True , annot = True, square = True,fmt = '.1f',annot_kws = {'size':8},cmap = 'Reds')
```
